DIABETES PREDICTION SYSTEM 

INTRODUCTION:
A binary classification ML task that predicts whether a patient has diabetes (Outcome 1) or does not (Outcome 0) based on diagnostic measurements (e.g., Glucose, BMI, Age, etc.).

Application: This system is used in healthcare informatics to assist medical professionals in early screening and risk assessment for diabetes.

STEP 1: IMPORT libraries:

```python
# Import numerical computing library
import numpy as np

# Import data manipulation library
import pandas as pd

# Import pickle for saving and loading models
import pickle

# Import tools for splitting and evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm

# Import display libraries
from prettytable import PrettyTable
```

Step 2: LOAD SAMPLE DATA

```python
sample_data = pd.read_csv("diabetes_toy.csv")

# Print a header to indicate that sample data is being displayed
print("\n\nSample Data:")
print("============\n")

# Configure pandas to display all rows and columns (so nothing gets hidden with '...')
pd.set_option("display.max_rows", None, "display.max_columns", None)

# Print the complete dataset
print(sample_data)
```

step 3: understand sample data

```python
print("\n\nAttributes in Diabetes Data:")
print("============================\n")
print(sample_data.columns)

print("\n\nNumber of Instances in Toy Dataset:", sample_data["Age"].count())
print("========================================\n")
```

step 4&5: Label encoding

```python
# Inputs are already numeric; no Label Encoding required for features.

sample_data_encoded = sample_data.copy()

print("\n\nSample Data (Numeric):")
print("======================\n")
print(sample_data_encoded.head())

```

step 6 : splitting data

```python
# Split dataset into training (80%) and testing (20%)
# random_state=0 ensures reproducibility (same split every time)
# shuffle=False keeps the original order of rows (no shuffling before splitting)
training_data_encoded, testing_data_encoded = train_test_split(
    sample_data_encoded,
    test_size=0.2,
    random_state=0,
    shuffle=False
)

# Save the training and testing datasets to CSV files
training_data_encoded.to_csv(r'training-data-encoded.csv', index=False, header=True)
testing_data_encoded.to_csv(r'testing-data-encoded.csv', index=False, header=True)

# Print training dataset
print("\n\nTraining Data:")
print("==============\n")
print(training_data_encoded)

# Print testing dataset
print("\n\nTesting Data:")
print("==============\n")
print(testing_data_encoded)

```

splitting input vectors and outputs

```python
# X = Features, y = Outcome
input_vector_train = training_data.iloc[:, :-1]
output_label_train = training_data.iloc[:, -1]

print("\n\nFeature Names:")
print(input_vector_train.columns.tolist())
```

TRAIN the supoort vector classifier

```python
print("\n\nTraining the Support Vector Classifier...")
# We use a linear kernel for medical numeric data
svc_model = svm.SVC(kernel='rbf', C=10, gamma=0.1, random_state=0)
svc_model.fit(input_vector_train, np.ravel(output_label_train))

print("\nTrained Model Parameters:")
print(svc_model)
```

save the trained model

```python
# Save for deployment
pickle.dump(svc_model, open('diabetes_trained_model.pkl', 'wb'))
print("Model saved as diabetes_trained_model.pkl")
```

testing and accuracy

```python
# Load model
model = pickle.load(open('diabetes_trained_model.pkl', 'rb'))

# Prepare test data
input_vector_test = testing_data.iloc[:, :-1]
actual_labels = testing_data.iloc[:, -1]

# Predict
predictions = model.predict(input_vector_test)

# Calculate Accuracy
score = accuracy_score(actual_labels, predictions)
print(f"\n\nModel Accuracy Score: {round(score * 100, 2)}%")
```

Application phase:
1. take input from user

```python
# Take Pregnancies input
preg_input = input("\nPlease enter Number of Pregnancies: ").strip()

# Take Glucose input
glucose_input = input("\nPlease enter Glucose level (mg/dL): ").strip()

# Take Blood Pressure input
bp_input = input("\nPlease enter Blood Pressure (mm Hg): ").strip()

# Take Skin Thickness input
skin_input = input("\nPlease enter Skin Thickness (mm): ").strip()

# Take Insulin input
insulin_input = input("\nPlease enter Insulin level (mu U/ml): ").strip()

# Take BMI input
bmi_input = input("\nPlease enter BMI (e.g., 26.5): ").strip()

# Take Diabetes Pedigree Function input
pedigree_input = input("\nPlease enter Diabetes Pedigree Function (e.g., 0.45): ").strip()

# Take Age input
age_input = input("\nPlease enter Age (years): ").strip()


print("\nUser Inputs Captured Successfully:")
print("====================================")
print(f"Pregnancies: {preg_input}")
print(f"Glucose: {glucose_input}")
print(f"Blood Pressure: {bp_input}")
print(f"Skin Thickness: {skin_input}")
print(f"Insulin: {insulin_input}")
print(f"BMI: {bmi_input}")
print(f"Pedigree: {pedigree_input}")
print(f"Age: {age_input}")
```

Convert User Input into Feature Vector

```python

user_input_df = pd.DataFrame({
    'Pregnancies': [float(preg_input)],
    'Glucose': [float(glucose_input)],
    'BloodPressure': [float(bp_input)],
    'SkinThickness': [float(skin_input)],
    'Insulin': [float(insulin_input)],
    'BMI': [float(bmi_input)],
    'DiabetesPedigreeFunction': [float(pedigree_input)],
    'Age': [float(age_input)]
})

# Print the feature vector created from user input
print("\n\nUser Input Feature Vector (Diabetes):")
print("=====================================\n")
print(user_input_df)
```

Load the saved model

```python
model = pickle.load(open('diabetes_trained_model.pkl', 'rb'))

print("\nModel Loaded Successfully!")
print("==========================")
print(model)
```

Model Prediction

```python

predicted_outcome = model.predict(user_input_df)


if predicted_outcome[0] == 1: 
    prediction = "DIABETIC"
else:
    prediction = "NON-DIABETIC"


pretty_table = PrettyTable()
pretty_table.add_column("    Prediction Result  ", [prediction])


print(pretty_table)
```
